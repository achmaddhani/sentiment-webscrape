**3.2 Features Creations**

**3.2.1 Adding Column 'Hashtags'**

```python
# Mencari dan mengekstrak hashtag dari kolom full_text
data['hashtag'] = data['Tweets'].str.findall(r'#\w+')
data
```

**3.2.2 Adding Columns Sentiment Label**

```python
# Import filter warnings module to ignore the warnings
import warnings
warnings.filterwarnings("ignore")
# Importing the pipeline module from transformers package
from transformers import pipeline
```

```python
# Generating the model obtained from indonesian roberta base sentiment classifier
pretrained_name = "w11wo/indonesian-roberta-base-sentiment-classifier"

nlp = pipeline(
    "sentiment-analysis",
    model=pretrained_name,
    tokenizer=pretrained_name
)
```

```python
# Categorizing the Sentiment Labels
def sentiment_result(text):
    result= nlp(text)
    if result[0]['label'] == 'positive':
        return 'Positive'
    elif result[0]['label'] == 'negative':
        return 'Negative'
    else:
        return 'Neutral'
```

```python
# Testing the NLP Model
test = nlp('Indonesia adalah negara berkembang')
```

```python
# Displaying the results 
test[0]['label']
```

```python
data['TweetSentiments']= data['Tweets'].apply(sentiment_result)
data
```
